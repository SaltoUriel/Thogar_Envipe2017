```python
import pandas as pd
import numpy as np
```

```python
path = 'C:/Users/Uriel/Documents/DELFIN/Casas de Seguridad/thogar_envipe2013/thogar.csv'
df = pd.read_csv(path, encoding='latin', low_memory=False)
```

```python
df
```

El primer paso, sera ver si las columnas contienen valores nulos. Lo cual se realizara de la siguiente forma

```python
df.UPM.isnull().value_counts()
```

En caso de que alguna columna tenga valores nulos, se remplazan de la siguiente forma

```python
df['R_SEL'].fillna('1', inplace = True)
```

Se puede ver que la tabla no esta muy sucia, por lo que solo le haremos los cambios necesarios

Se modificaran las columnas 'RESUL_H' y 'DOM' debido a que sus valores son categoricos, por lo que se remplazaran por numeros.

Columna RESUL_H; Valores originales : A, B, C, D, E Valores nuevos : 1, 2, 3, 4, 5

Columna DOMINIO; Valores originales : U, C, R Valores nuevos : 1, 2, 3

Remplaso de valores de la columna RESUL_H

```python
df.RESUL_H=  df.RESUL_H.replace( 'A', "1", regex = True  )

df.RESUL_H=  df.RESUL_H.replace( 'B', "2", regex = True  )

df.RESUL_H=  df.RESUL_H.replace( 'C', "3", regex = True  )

df.RESUL_H=  df.RESUL_H.replace( 'D', "4", regex = True  )

df.RESUL_H=  df.RESUL_H.replace( 'E', "5", regex = True  )
```

Remplaso de valores de la columna DOMINIO

```python
df.DOM=  df.DOM.replace( 'U', "1", regex = True  )

df.DOM=  df.DOM.replace( 'C', "2", regex = True  )

df.DOM=  df.DOM.replace( 'R', "3", regex = True  )
```
