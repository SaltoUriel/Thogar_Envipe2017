```python
import pandas as pd
import numpy as np
```

Importacion de la tabla thogar_2016 al data frame

```python
path = 'C:/Users/Uriel/Documents/DELFIN/Casas de Seguridad/thogar_envipe2016/conjunto_de_datos/thogar_envipe2016.csv'
df = pd.read_csv(path, encoding='latin', low_memory=False)
```

```python
df
```

 Se limpiaran las columnas 'ID_HOG', 'RESUL_H','N_INF', 'R_SEL' 'DOMINIO' ya que cuentan con un caracter invalido

```python
df.ID_HOG=  df.ID_HOG.replace( '\r', "", regex = True  )
```

```python
df.RESUL_H=  df.RESUL_H.replace( '\r', "", regex = True  )
```

```python
df.N_INF=  df.N_INF.replace( '\r', "", regex = True  )
```

```python
df.R_SEL=  df.R_SEL.replace( '\r', "", regex = True  )
```

```python
df.DOMINIO=  df.DOMINIO.replace( '\r', "", regex = True  )
```

```python
df
```

La columna 'FAC_HOG_AM' se eliminara debido a que mas de el 50% de sus valores son nulos. 

```python
del df['FAC_HOG_AM']
```

Se modificaran las columnas 'RESUL_H' y 'DOMINIO' debido a que sus valores son categoricos, por lo que se remplazaran por numeros.

Columna RESUL_H; Valores originales : A, B, C, D Valores nuevos : 1, 2, 3, 4

Columna DOMINIO; Valores originales : U, C, R Valores nuevos : 1, 2, 3

Remplaso de valores de la columna RESUL_H

```python
df.RESUL_H=  df.RESUL_H.replace( 'A', "1", regex = True  )

df.RESUL_H=  df.RESUL_H.replace( 'B', "2", regex = True  )

df.RESUL_H=  df.RESUL_H.replace( 'C', "3", regex = True  )

df.RESUL_H=  df.RESUL_H.replace( 'D', "4", regex = True  )
```

Remplaso de valores de la columna DOMINIO

```python
df.DOMINIO=  df.DOMINIO.replace( 'U', "1", regex = True  )

df.DOMINIO=  df.DOMINIO.replace( 'C', "2", regex = True  )

df.DOMINIO=  df.DOMINIO.replace( 'R', "3", regex = True  )
```

```python
df
```

Ya limpia la tabla, se exportara a .csv 

```python
df.to_csv('C:/Users/Uriel/Documents/DELFIN/Casas de Seguridad/thogar_envipe2016/conjunto_de_datos/thogar_envipe2016_clean.csv')
```
