Columna DOMINIO; Valores originales : U, C, R Valores nuevos : 1, 2, 3

Remplaso de valores de la columna RESUL_H

```python
df.RESUL_H=  df.RESUL_H.replace( 'A', "1", regex = True  )

df.RESUL_H=  df.RESUL_H.replace( 'B', "2", regex = True  )

df.RESUL_H=  df.RESUL_H.replace( 'C', "3", regex = True  )

df.RESUL_H=  df.RESUL_H.replace( 'D', "4", regex = True  )
```

```python
df.DOMINIO=  df.DOMINIO.replace( 'U', "1", regex = True  )

df.DOMINIO=  df.DOMINIO.replace( 'C', "2", regex = True  )

df.DOMINIO=  df.DOMINIO.replace( 'R', "3", regex = True  )
```
